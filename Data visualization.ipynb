## using a bar chart for a histogram

```python
from collections import Counter
scores = [95, 81, 91, 87, 70, 0, 85, 88, 82, 100, 67, 73, 72, 7, 98, 55, 59]
histogram = Counter(min(score // 10*10, 90)
                    for score in scores)
plt.bar([x + 5 for x in histogram.keys()],     # shift bars right by 5
        histogram.values(),                    # give each bar its correct height
        10,                                    # give each bar a width of 10
        edgecolor=(0,0,0))                     # black edges for each bar
plt.axis([-5,105,0,5])
plt.xlabel("Scores in Decile")
plt.xticks([10 * i for i in range(11)])
plt.ylabel("Scores Recorded")
plt.title("Distribution of Math Scores")
plt.show()
```

## line charts

```python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256, 512]
bias_squared = [512, 256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [a + b for a,b in zip(variance, bias_squared)]
xs = [x for x, _ in enumerate(variance)]
```

```python
plt.plot(xs, variance, 'g-', label='variance')
plt.plot(xs, bias_squared, 'r-', label='bias^2')
plt.plot(xs, total_error, 'b:', label='total error')

plt.legend(loc=9)
plt.xlabel('model complexity')
plt.xticks([])
plt.title("Bias-Variance Tradeoff")
plt.show()
```

## scatterplot

```python
friends = [70, 62, 65, 72, 63, 71, 64, 60, 64, 67, 69]
tweets = [200, 115, 170, 205, 120, 220, 130, 105, 145, 190, 199]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k']
```

```python
plt.scatter(friends, tweets)
for label, friend_count, tweet_count in zip(labels, friends, tweets):
    plt.annotate(label,
                 xy = (friend_count, tweet_count),
                 xytext = (5, -5), 
                 textcoords = 'offset points')

plt.title("Daily Tweets vs Number of Friends")
plt.xlabel("# of friends")
plt.ylabel("daily tweets spent on the site")
plt.show()

```
